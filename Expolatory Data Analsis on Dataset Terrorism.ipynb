```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
df[['INT_LOG', 'INT_IDEO', 'INT_MISC', 'INT_ANY']].boxplot()
plt.title('INT_LOG, INT_IDEO, INT_MISC,INT_ANY')
plt.show()
```

```python
sns.countplot(data=df, x='region', hue='country_txt')
plt.title('Count Plot by region and country')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df[['country', 'region', 'related', 'iyear']].boxplot()
plt.title('country, region, related, iyear')
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
gdf = gpd.GeoDataFrame(
    df, 
    geometry=gpd.points_from_xy(df.longitude, df.latitude)
)
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = world.plot(figsize=(10, 6), color='white', edgecolor='black')
gdf.plot(ax=ax, color='red', markersize=5)
plt.title('Terrorism Scatter Map')
plt.show()
```
